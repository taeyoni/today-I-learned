```python
#boxplot 으로 전체 변수 시각화하기
#개별값 보기 어려움 따라서 스케일 조정 등을 통해 개별로 보기 편하게 해야함.
sns.boxplot(data=df)
```

```python
#2개 이상의 수치 비교
#scatterplot 을 통해 2개의 수치변수 비교하기
sns.scatterplot(data=df, x="weight", y="mpg")
```

```python
#회귀 시각화
sns.regplot(data=df, x="weight", y="mpg")
```

```python
#잔차 시각화
#회귀선의 잔차 시각화
sns.residplot(data=df, x="weight", y="mpg")
```

```python
#lmplot 통해 범주값에 따라 색상 그리기
sns.lmplot(data=df, x="weight", y="mpg", hue="origin")
```

```python
#lmplot 통해 범주값에 따라 색상, 서브플롯 그리기(by columns 추가)
sns.lmplot(data=df, x="weight", y="mpg", hue="origin", col="origin")
```

```python
#lmplot 통해 범주값에 따라 색상, 서브플롯 그리기(by columns 추가), 회귀선 길게 그리기(truncate=Fales)
sns.lmplot(data=df, x="weight", y="mpg", hue="origin", col="origin", truncate=False)
```

```python
#jointplot으로 scatterplot 그려서 수치 비교
sns.jointplot(data=df, x="weight", y="mpg")
```

```python
#jointplot으로 scatterplot 그려서 수치 비교// kind 통해서 다양한 형식으로 그릴 수 있다.
sns.jointplot(data=df, x="weight", y="mpg", kind="kde" )
```

```python
#pairplot
#시간이 오래 걸려 샘플 추출 후  샘플 수 늘려가며 그리기
df_sample = df.sample(100)
df_sample.shape
```
