```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dart_b-data/toyproject/renewable_energy.csv', encoding='euc-kr')
df.head()
```

```python
df.drop(columns=['태양광발전량(MWh)'], inplace=True)
df.head()
```

```python
# '지역'별로 그룹화하고 다양한 집계 통계를 계산
grouped_df = df.groupby('지역')[['풍력발전량(MWh)']].agg(['sum', 'mean', 'max', 'min'])
grouped_df.head(15)
```
