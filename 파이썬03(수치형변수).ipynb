```python
#unique값 보기
df["mpg"].unique()
```

```python
#히스토그램 그리기
#bins(막대개수)
_ = df.hist(figsize=(10, 8), bins=50)
```

```python
#skew로 왜도 구하기
#왜도 = 확률분포의 비대칭성을 나타내는 지표
#평균 = 중앙값 왜도 0
#df.skew() 에러가 왜 뜨는걸까....
```

```python
#첨도 구하기
#첨도 = 확률분포의 뾰족한 정도를 나타내는 척도
#df.kurt() 에러가 왜 뜨는걸까....
```

```python
#displot통해 히스토그램과 kdeplot 그리기
sns.displot(data=df, x="mpg", kde=True, rug=True)
```

```python
#kdeplot, rugplot으로 밀도함수 표현하기
#밑에 러그처럼 어디에 많이 몰려있는지 확인가능
sns.kdeplot(data=df, x="mpg", shade=True, cut=3)
sns.rugplot(data=df, x="mpg")
```

```python
#mpg 값에 대해 agg로 skew, kurt 값 구하기
df["mpg"].agg(["skew", "kurt"])
```

```python
#boxplot으로 mpg의 사분위 수 표현하기
sns.boxplot(data=df, x="mpg")
```

```python
#describe로 mpg의 기술통계 값 구하기
df["mpg"].describe()
```

```python
#violinplot으로 mpg 값 더 자세히 보기
#boxplot의 단점 보완(박스플롯 마주 봄)
sns.violinplot(data=df, x="mpg")
```

```python
#boxplot 으로 전체 변수 시각화하기
#개별값 보기 어려움 따라서 스케일 조정 등을 통해 개별로 보기 편하게 해야함.
sns.boxplot(data=df)
```

```python
# violinplot 으로 전체 변수 시각화하기
sns.violinplot(data=df)
```
